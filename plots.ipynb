```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('longest_chain_data.csv')

fig, ax = plt.subplots(figsize=(8,6))
# plt.rcParams['figure.dpi'] = 200
plt.rcParams.update({'font.size': 14})
fig.set_facecolor('white')

# fig.subplots_adjust(top=0.8)
# fig.suptitle('Fruitchain longest chain rule')

ax.grid()

# ax.set_title('Fruitchain longest chain rule')

ax.set_xlabel('Mining Power of Attacker [%]')

ax.set_ylabel('Reward of Attacker [%]')

ax.plot(df['selfish_miner_power'].to_numpy(), df['profit'].to_numpy(), 'r', linewidth=3.0, label='Attacker')
ax.plot(df['selfish_miner_power'].to_numpy(), df['selfish_miner_power'].to_numpy(), '--', color='gray', linewidth=2.0, label='Baseline')
ax.legend()

plt.show()
fig.savefig('fruitchain_longest_chain.pdf')
```
